```python
import pandas as pd
import numpy as np
import datetime 

ahora=datetime.datetime.now()
print(ahora.strftime('%d/%m/%Y/%H:%M:%S'))

file='student-mat.csv'
def open_(file):
    data=pd.read_csv(file)
    return(data)
df=open_(file)
df.head(10)

df = data = pd.read_csv(file,sep=';')
df = df.drop(['famsize', 'Medu','Fedu','Mjob','Fjob','reason'], axis=1)
df = df.drop(['schoolsup','famsup', 'activities','nursery','higher','romantic'], axis=1)
df = df.drop(['famrel','freetime', 'goout','Dalc','Walc'], axis=1)

df ['extra']=np.where((df['G3']>15),'0','1')
df ['Aproved']=np.where((df['absences']<15)&(df['G3']<10),'0','1')

df
```

```python
df.to_csv('resultado-mat.csv')
```
